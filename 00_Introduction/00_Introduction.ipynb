## What is Python?

<img src="images/monty_python.jpg" width="400">

* A scripted high-level programming language created by [Guido Van Rossum](https://en.wikipedia.org/wiki/Guido_van_Rossum) and named after [Monty Python's Flying Circus](https://en.wikipedia.org/wiki/Monty_Python%27s_Flying_Circus) 

* Easy-to-use, versatile and with an emphasise on readability 

* It has a minimalistic English-like syntax, relying on indentation instead of curly brackets, semicolons etc.

## Why Python?

Here's the [TIOBE index](http://www.tiobe.com/tiobe-index/) (a measure of the popularity of programming languages):

<img src="images/tiobe.png" width=700>

Python is becoming the de facto standard for exploratory and interactive scientific research. Here are some reasons for its popularity:

* It is free! No licence costs

* Runs on all platforms (Mac, Windows, Linux)

* Because of it's ease of programming (no neeed to worry about allocating memory or defining variable types), Python minimises development effort and maintainability

* A huge number of [libraries](https://pypi.python.org/pypi), and an active and diverse [community](https://www.python.org/community/) behind them  

* Python can "glue" together functions written in C/C++ and Fortran to reduce computational effort (we can also call R and MATLAB functions from within Python)

* Compared to other high-level scientific languages such as MATLAB and R, Python offers a much wider range of additional functionality (e.g [web development](https://www.djangoproject.com/)), hence the nickname “the swiss army knife” of programming languages. 

**Horses for courses**  
Python is no programming silver bullet. Your application will ultimately dictate the tool. For example MATLAB excels at interfacing with hardware, e.g generating [hardware description language (HDL) code](https://uk.mathworks.com/products/hdl-coder.html) to configure an integrated circuit board or connecting to a [data acquisition card](https://uk.mathworks.com/products/daq.html). Similarly if your focus is data wrangling and statistics then R is a great choice.

## Running Python code


### 1. No frills Python interpreter
Once Python is installed on your machine, you can type `python` in your terminal window to invoke the interpreter. You will see the symbol `>>>`. Any code that you type in is executed as Python code once you press enter.  

<img src="images/python_interpreter.png">

Alternatively if your code is written in a text file, e.g `my_script.py` (we typically use the `.py` extension to highlight that it's Python code). Then the script can be run through the command line as follows:

`python my_script.py`

### 2. IPython
IPython is an interactive shell (similar to R Console), adding "frills" to the vanilla interpreter. In particular it:

* supports syntax highlighting (making it easier to read code)
* provides tab auto-completion (minimises typeos and listing available functions etc.)  

<img src="images/ipython.png">

The IPyton interactive shell is what powers Jupyter in the background (i.e code that I write in these notes is interpreted and executed by the IPython shell)

### 3. Spyder

## Python 2 vs 3

Python 2.7.X and Python 3.X are the two main versions of Python. In a nutshell, [Python 2.x is legacy, Python 3.x is the present and future of the language](https://wiki.python.org/moin/Python2orPython3). However, because not all Python 3.X code is backwards-compatible, there are still numerous modules that are written in Python 2.7.X, so it's important to be aware of [key differences](http://sebastianraschka.com/Articles/2014_python_2_3_key_diff.html). In this workshop we will use Python 3.X, which is the language being actively developed. Nevertheless, when it comes to scientific computing, you will find that the differences between Python 2 and 3 are small.

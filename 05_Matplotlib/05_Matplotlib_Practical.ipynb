# Basic plots

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate sinusoidal data
x = np.linspace(0, 4*np.pi, 100)
y = np.sin(x)

# Plot sinusoidal curve
plt.plot(x, y, color="blue")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.title("My first Python plot")
```

```python
plt.subplot(2, 1, 1) # 2 rows, 1 column, first plot
plt.plot(x, np.sin(x), color="b", label="Sine")
plt.ylabel("Amplitude")
plt.legend(loc="upper right")

plt.subplot(2, 1, 2) # 2 rows, 1 column, second plot
plt.plot(x, np.cos(x), color="r", label="Cosine")
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.legend(loc="upper right")
```

# Statistical plots

```python
# Generate 1000 normally distributed numbers
x1 = np.random.randn(1000)
x2 = np.random.randn(1000)

# 1D histogram (x1)
plt.subplot(1, 2, 1)
plt.hist(x1, bins=20)

# 2D histogram (x1 vs x2)
plt.subplot(1, 2, 2)
plt.hist2d(x1, x2, bins=20) 
plt.colorbar() # display colourmap
```

```python
# Generate four sets of random numbers
data = [np.random.randn(1000) for i in range(4)]

# Boxplot
plt.subplot(1, 2, 1)
plt.boxplot(data)

# Violin plot
plt.subplot(1, 2, 2)
plt.violinplot(data)
```

# 3D plots
Let us generate some artificial 3D data. In this case we will plot the exponential function

$$
z = e^{-x^2 - y^2}
$$

First we need to create a grid/mesh, a 2D `Numpy` array, on which $z$ is computed on.

```python
# Generate some x, y
x = np.arange(-2.5, 2.5, 0.1)
y = np.arange(-2.5, 2.5, 0.1)
x, y = np.meshgrid(x, y) # creates a mesh/grid
x, y
```

Now we can compute $z$ for **each** combination of $x$ and $y$. Note how through `Numpy`'s vectorisation process we do not need to loop through all elements of $x$ and all elements of $y$.

```python
# Compute function for each combination of x and y
z = np.exp(-x**2 - y**2)
z
```

Now we have three variables ($x, y ,z$) represented as a 2D `Numpy` matrix. To plot in 3D, we first set the figure's axis to 3D then plot the data.

```python
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm

# Create figure amenable for 3D plotting
hFig = plt.figure() # hFig - a handle to the figure object
hAx = hFig.gca(projection="3d") # hAx - a handle to the axis object

# Plot the surface
hSurf = hAx.plot_surface(x, y, z, cmap=cm.inferno) # hSurf - a handle to the 3D surface object
plt.colorbar(hSurf, shrink=.5)
```

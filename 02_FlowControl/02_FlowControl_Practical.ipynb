# Motivation
* Executing code one line at a time is useful but limiting
* **Flow control** commands lets us dictate the **order** in which commands are run:
    1. **If-else**: to change what commands are executed under certain conditions
    2. **For loops**: to repeat the same thing N times
    3. **While loops**: to repeat the same thing until a specific condition is met

# If-else
The if-else statement let us execute commands *only* if certain conditions are met. Consider the simple example of printing whether an integer is positive, negative or zero.  

```python
x = 0

if x > 0:
    print("x is positive") 
elif x < 0: # read as "else if"
    print("x is negative") 
else:
    print("x is zero")
```

We can have as many `elif` (read as else if) as we want. Note however that once a condition is met, the program will exit the `if...elif...else` statement.

```python
x = 1

if x > 0:
    print("x is positive") 
elif x >= 1:
    print("x is greater than one") 
elif x < 0:
    print("x is negative") 
else:
    print("x is zero")
```

Note how the "x is greater than one" is not executed because the program exits the `if...elif...else` statement once the first condition (`x > 0`) is met.

**NOTE**: Unlike most other programming languages we don't need curly brackets { } to encapsulate multiple commands after the `if`, **indentation** is everything in Python!

```python
x = -2

if x > 0:
    print("x = {0}".format(x)) # no need for { }
    print("x is positive") 
elif x < 0:
    print("x = {0}".format(x))
    print("x is negative") 
else:
    print("x = {0}".format(x))
    print("x is zero")
```

# For loops
A common task in scientific computing is to perform the same operation several times. Consider the (not very useful) example of wanting to print "Hello world" five times. To do this we need to create an *iterator* (a sequence of integers) using the `range` command.

```python
for i in range(5):
    print("{0}: Hello world".format(i))
```

Note how the iterate `i` changes as we go through the loop; `i=0`, then `i=1`, until `i=4` (remember `range(5)` creates the sequence 0, 1, 2, 3, 4).

We can also iterate over a list. Suppose we have a list of strings and for each string we want to compute its length (using the `len()` in-built function) and print it out in upper case (using the `.upper()` method of strings).

```python
countries = ["Uganda", "United Kingdom", "Kenya", "Germany"]

for country in countries:
    NLength = len(country) # compute the length of the string
    countryUpper = country.upper() # convert the string to upper case
    print("{0} contains {1} characters".format(countryUpper, NLength))
```

So for the first iteration, `country = "Uganda"`, then `country = "United Kingdom"` etc..

# While loops
Similar to for loops, but instead of specifying the number of loops via an iterator, the program keeps on looping until a specific condition is met. For example we can use a while loop to print the integers 10 to 1.

```python
x = 10
while x > 0:
    print(x) 
    x = x - 1
```

As long as x is positive keep on executing the commands `print(x)` and `x = x - 1`. Note that we could have achieved the same thing using a for loop

```python
for x in range(10, 0, -1):
    print(x)
```

In general:    
1. Use `for loops` over `while loops` where possible
2. Ensure that the while condition evaluates to False at some point to avoid an infinite loop

# List comprehensions
Lists are a popular data type in Python due to their versatility. List comprehensions are simply an optimized and readible method for creating a list using a "shorthand" for loop. Recall the previous example where we converted a list of country names to upper case. Suppose we wanted to store the list of countries in upper case in another variable. To do this using for loops we need to create an empty list and then sequentially append the upper cased name to it.

```python
countries = ["Uganda", "United Kingdom", "Kenya", "Germany"]

countriesUpper = []
for country in countries:
    countriesUpper.append(country.upper()) # the .append() list method adds a new element to the list

print(countriesUpper)
```

Alternatively we can achieve the above in a conveniently one line command called a list comprehension

```python
countriesUpper = [country.upper() for country in countries]

print(countriesUpper)
```

We can also take this further and introduce an `if` clause to avoid including "Kenya" for example from the list of countries.

```python
countriesUpper = [country.upper() for country in countries if country != "Kenya"]

print(countriesUpper)
```

# Enumerate
Looping over lists is very common. However, sometimes we also want to keep track of how many times we have looped over (i.e the index of the iteration, 0, 1, 2, etc.). Using the countries example, let us print the country in upper case next to the index using a normal for loop. In this case we need to keep track of an index ourselves.

```python
countries = ["Uganda", "United Kingdom", "Kenya", "Germany"]

i = 1 # index counter
for country in countries:
    print("{0}. {1}\n".format(i, country.upper()))
    i = i + 1 # update index counter
```

**Enumerate** is a built-in function that lets us access the iterable element but also its index.

```python
for i, country in enumerate(countries):
    print("{0}. {1}\n".format(i + 1, country.upper())) # we need the + 1 as Python starts from 0
```
